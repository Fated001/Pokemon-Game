```python
import random as rd
import math as math
class Pokemon:


    def __init__(self, name):
        self.name = name
        self.types = POKEMONS[name]['Type']
        self.hp = POKEMONS[name]['HP']
        self.attack = POKEMONS[name]['Attack']
        self.defense = POKEMONS[name]['Defense']
        self.speed = POKEMONS[name]['Speed']
        self.exp = POKEMONS[name]['Exp']
        self.moves = POKEMONS[name]['Moves']
        self.my_exp = 0

    @property
    def level(self):
        if self.my_exp == 0:
            return 1
        else:
            n = math.floor(self.my_exp**(1/3))
            return int(n)


    def attack_opponent(self, opponent, move):
        n = rd.randint(0,265)
        if n <= self.speed:
            crit = 2
        else:
            crit = 1
        m = rd.randint(85,100)/100
        for type_, moves in MOVES.items():
            if move in moves:
                move_type = type_
        effective = 1
        for kind in opponent.types:
            if kind in SUPER_EFFECTIVE[move_type]:
                num = 2
            elif kind in NOT_VERY_EFFECTIVE[move_type]:
                num = 0.5
            else:
                num = 1
            effective = effective*num
        modifier = m*crit*effective
        power = POWERS[move]
        damage = (((((2*self.level)/5)+2)*(self.attack/self.defense)*power)/50)*modifier
        print("\n{:20}You attacked {} with {}! They lost {:0.3f} hp!".format('', opponent.name, move, damage))
        opponent.hp -= damage


    def take_damage(self, opponent, move):
        n = rd.randint(0,266)
        if n <= opponent.speed:
            crit = 2
        else:
            crit = 1
        m = rd.randint(85,101)/100
        for key, value in MOVES.items():
            if move in value:
                move_type = key
        effective = 1
        for kind in self.types:
            if kind in SUPER_EFFECTIVE[move_type]:
                num = 2
            elif kind in NOT_VERY_EFFECTIVE[move_type]:
                num = 0.5
            else:
                num = 1
            effective = effective*num
        modifier = m*crit*effective
        power = POWERS[move]
        damage = (((((2*opponent.level)/5)+2)*(opponent.attack/opponent.defense)*power)/50)*modifier
        print("\n{:>45} attacked you with {}! You took {:0.3f} damage.".format(opponent.name, move, damage))
        self.hp -= damage

    def __str__(self):
        return f'''    Name: {self.name}
    Type/s: {self.types}
    HP: {self.hp}
    Defense: {self.defense}
    Attack: {self.attack}
    Speed: {self.speed}
    Experience Gained: {self.exp}
    Moves: {self.moves}
    Your Exp: {self.my_exp}'''
```
