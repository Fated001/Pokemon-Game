```python
# %%pycodestyle
import random as rand
import math
import time


def input_number_range(num1, num2):
    while True:
        choice = input()
        try:
            choice = int(choice)
            if num1 <= choice <= num2:
                break
            else:
                print("Invalid input. Please choose %d-%d"%(num1, num2))
        except:
            print("Invalid input. Please choose %d-%d"%(num1, num2))
    return choice


class Pokemon:
    def __init__(self, name, type_, HP, Atk, Def, Spd, Exp, Moves):
        self.name = name
        self.type_ = type_
        self.fullHP = HP
        self.HP = HP
        self.Atk = Atk
        self.Def = Def
        self.Spd = Spd
        self.Exp = Exp
        self.Moves = Moves

    @property
    def Level(self):
        """Level of the pokemon"""
        if self.Exp == 0:
            return 1
        else:
            return math.floor(self.Exp**(1/3))

    def stat_increase(self):
        """Increase stats after leveling up"""
        self.fullHP += (1/50)*CHARACTERS[self.name]["HP"]
        self.Atk += (1/50)*CHARACTERS[self.name]["Attack"]
        self.Def += (1/50)*CHARACTERS[self.name]["Defense"]
        self.Spd += (1/50)*CHARACTERS[self.name]["Speed"]

    def level_increase(self, exp_gained):
        '''Display Level increased text.'''
        old_level = self.Level
        self.Exp += exp_gained
        while old_level != self.Level:
            print(f"Your {self.name} has leveled up from {old_level} -> {old_level+1}")
            time.sleep(0.5)
            old_level += 1
            self.stat_increase()

    def effective(self, move, defending):
        """Calculate the effectiveness of the move against another pokemon"""
#         print(self.name, move, defending.type_)
        mod = 1
        for t in MOVES_DICTIONARY[move]["super effective against"]:
            if t in defending.type_:
                mod *= 2
        for t in MOVES_DICTIONARY[move]["not very effective against"]:
            if t in defending.type_:
                mod /= 2
        return mod

    def crit(self):
        """Determine the critical hit"""
        p = self.Spd/512
        num = rand.random()
        if num < p:
            return 2
        else:
            return 1

    def battle(self, B):
        """Battle Interface (w/Computer)"""
        turn = 0
        while self.HP > 0 and B.HP > 0:
            print()
            print(f"Your HP: {math.ceil(self.HP)}")
            print(f"Wild {B.name}\'s HP: {math.ceil(B.HP)}")
            if turn == 0:
#                 Player's turn
                print(f"What would {self.name} do? (1-{len(self.Moves)})")
                for i in range(1, len(self.Moves)+1):
                    print(str(i)+'. '+self.Moves[i-1])
                att = input_number_range(1, len(self.Moves))
                print(f"You used {self.Moves[att-1]}")
                time.sleep(2)
                dmg = self.calculate_damage(B, self.Moves[att-1])
#                 print(dmg, self.Moves[att-1])
                mod = self.effective(self.Moves[att-1], B)
                cri = self.crit()
                dmg = dmg*mod*cri
#                 print(f"dmg = {dmg*mod*cri}")
                if mod > 1:
                    print("It's super effective!")
                elif mod < 1:
                    print("It's not very effective...")
                if cri == 2:
                    print("A critical hit!")
                print(f"Wild {B.name} took %d damage"%(math.ceil(B.HP)-math.ceil(B.HP-dmg)))
                time.sleep(2)
                B.HP -= dmg
            else:
#                 Computer's turn
                mx = (0, "")
#                 Caluculating the best moves
                for m in B.Moves:
                    dmg = B.calculate_damage(self, m) * B.effective(m, self)
                    if dmg > mx[0]:
                        mx = (dmg, m)
                print(f"{B.name} uses {mx[1]}")
                time.sleep(2)
                dmg = B.calculate_damage(self, mx[1])
                mod = B.effective(mx[1], self)
                cri = B.crit()
#                 print(dmg, mod, cri)
                dmg = dmg*mod*cri
#                 print(f"dmg = {dmg*mod*cri}")
                if mod > 1:
                    print("It's super effective!")
                elif mod < 1:
                    print("It's not very effective...")
                if cri == 2:
                    print("A critical hit!")
                print(f"You took %d damage"%(math.ceil(self.HP)-math.ceil(self.HP-dmg)))
                time.sleep(2)
                self.HP -= dmg
            turn = (turn+1) % 2
        if B.HP > 0:
            print("Your Pokemon fainted")
            return False
        else:
            print(f"You gained {B.Exp} EXP")
            time.sleep(1)
            self.level_increase(B.Exp)
            print(f"You added {B.name} to your collection.")
            time.sleep(1)
            B.HP = CHARACTERS[B.name]["HP"]
            self.HP = self.fullHP
            return True

    def calculate_damage(self, defending, move):
        """Calculate Raw Damage"""
        dmg = (2*self.Level/5+2)*MOVES_DICTIONARY[move]['power']*(self.Atk/defending.Def)
        return dmg/50
```
